# External Locations Audit

This notebook audits all External Locations in Unity Catalog.

## Purpose
- List External Locations
- Check assigned permissions (GRANTS)
- Validate access required for data pipelines

## Required access (PT Frozen Foods)
- RAW: READ FILES
- BRONZE: READ FILES, WRITE FILES

## Notes
- This notebook is read-only (no changes applied)
- Permissions are managed in the `03_permissions` notebook

```python
# Audit grants for all external locations in the environment
external_locations = spark.sql("SHOW EXTERNAL LOCATIONS").collect()

for el in external_locations:
    name = el['name']
    print(f"\nExternal Location: {name}")
    display(spark.sql(f"SHOW GRANTS ON EXTERNAL LOCATION `{name}`"))
```
